```python
plt.imshow(img);
```

the color is not RGB because we import image through opencv and plot the image through mathplotlib. \
the way matplotlib and opencv read the image is different. \
`mathplotlib` ---> RGB = Red, Green, Blue \
`OpenCV` ---> RGB = BGR = Blue, Green, Red \

#### let's fix this problem

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img);
```
